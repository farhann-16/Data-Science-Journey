**Goal:** Explore pairplot and scatterplot with regression lines. 

# 1. Pairplot

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
sns.pairplot(tips, diag_kind="kde", hue="sex")
plt.show()
```

# 2. Scatterplot with Regression

```python
sns.lmplot(x="total_bill", y="tip", data=tips, hue="sex", aspect=1.2)
plt.title("Scatterplot with Regression")
plt.show()
```
