#  1. Multiple Line Plots

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 2, 3, 4, 5]

plt.plot(x, y1, label="Double", color="blue", marker="o")
plt.plot(x, y2, label="Single", color="red", linestyle="--")
plt.legend()
plt.title("Multiple Line Plots")
plt.show()
```

# 2. Subplots

```python
fig, axs = plt.subplots(1, 2, figsize=(8,4))

# Left subplot
axs[0].bar(["A","B","C"], [3,7,5], color="orange")
axs[0].set_title("Bar Chart")

# Right subplot
axs[1].plot([1,2,3], [4,5,6], color="blue")
axs[1].set_title("Line Plot")

plt.tight_layout()
plt.show()
```

# 3. Styling

```python
plt.plot([1,2,3,4], [10,20,25,30], color="purple", linewidth=2, marker="s")
plt.title("Styled Line Plot", fontsize=14, color="green")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```
