## Logistic Regression (Classification)

**Goal:**  
Understand Logistic Regression for binary classification.  
We will cover:  
- What is Logistic Regression?  
- Train a model with Scikit-learn  
- Predict Pass/Fail based on study hours & attendance  

# 1. Import Libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

# 2. Create a Sample Dataset

```python
data = {
    "Study_Hours": [2, 4, 5, 6, 1, 7, 8, 3, 9, 10],
    "Attendance":  [60, 65, 70, 75, 50, 80, 85, 55, 90, 95],
    "Pass":        [0, 0, 1, 1, 0, 1, 1, 0, 1, 1]
}
df = pd.DataFrame(data)
df
```

# 3. Define Features (X) and Target (y)

```python
X = df[["Study_Hours", "Attendance"]]
y = df["Pass"]
```

# 4. Train/Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

# 5. Train Logistic Regression Model

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

# 6. Make Predictions

```python
y_pred = model.predict(X_test)
print("Predictions:", y_pred)
```

# 7. Model Evaluation

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

# 8. Try a New Prediction

```python
new_student = [[6, 80]]
result = model.predict(new_student)
print("Prediction (1=Pass, 0=Fail):", result[0])
```
