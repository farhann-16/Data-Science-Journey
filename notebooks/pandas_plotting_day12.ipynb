**Goal:** Use Pandas built-in `.plot()` for quick visualizations on a dataset with more variety.  

# 1. Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# 2. Create a Richer Dataset

```python
data = {
    "Year": [2018, 2019, 2020, 2021, 2022, 2023],
    "Sales": [100, 120, 90, 150, 180, 200],
    "Profit": [20, 25, 15, 35, 40, 50],
    "Customers": [200, 250, 220, 300, 350, 400]
}

df = pd.DataFrame(data)
df
```

# 3. Line Plot (Sales over Time)

```python
df.plot(x="Year", y="Sales", kind="line", marker="o", title="Sales Over Years")
plt.show()
```

# 4. Bar Plot (Profit by Year)

```python
df.plot(x="Year", y="Profit", kind="bar", color="orange", title="Profit by Year")
plt.show()
```

# 5. Multiple Line Plot (Sales vs Profit)

```python
df.plot(x="Year", y=["Sales", "Profit"], kind="line", marker="o", title="Sales vs Profit")
plt.show()
```

# 6. Scatter Plot (Customers vs Sales)

```python
df.plot(x="Customers", y="Sales", kind="scatter", color="green", title="Sales vs Customers")
plt.show()
```

# 7. Pie Chart (Share of Customers per Year)

```python
df.set_index("Year")["Customers"].plot(kind="pie", autopct="%1.1f%%", figsize=(5,5), title="Customer Share by Year")
plt.ylabel("")  # remove ylabel
plt.show()
```
