# 1. Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 2. Load Netflix Dataset

```python
df = pd.read_csv("netflix_titles.csv")
df.head()
```

# 3. Basic Info & Cleanup

```python
df.info()
df.isnull().sum()
```

# 4. Content Type Distribution
- Movies ðŸŽ¥ vs TV Shows ðŸ“º

```python
df["type"].value_counts().plot(kind="bar", color=["red","blue"], title="Movies vs TV Shows on Netflix")
plt.show()
```

# 5. Content Growth Over Years

```python
df["release_year"].value_counts().sort_index().plot(kind="line", marker="o", figsize=(10,5), title="Content Added Over Years")
plt.show()
```

# 6. Top 10 Countries by Content

```python
df["country"].value_counts().head(10).plot(kind="bar", figsize=(8,4), title="Top 10 Content-Producing Countries")
plt.show()
```
